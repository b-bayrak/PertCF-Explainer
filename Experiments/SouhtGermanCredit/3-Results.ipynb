```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

warnings.filterwarnings("ignore")
```

-----

```python
res_list_names = ['coef', 'num_iter', 'dissimilarity', 'sparsity', 'instability', 'time']
num_iter = [2,3,4,5,10,20]

df = pd.DataFrame([],columns=res_list_names) 

for i in num_iter:
    df = df.append(pd.read_pickle('./Output/exp1_coef_1-750_numIter_'+ str(i) +'.pkl'))

df.reset_index(inplace=True)
df=df.drop(['index'],axis=1)
```

```python
sns.set_style("ticks")
sns.set_context("talk")

sns.lineplot(x = "coef", y = "dissimilarity", data = df, style = "num_iter", hue = "num_iter", palette = 'Paired')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.savefig('./Output/img/dissimilarity-coef.png', dpi=1200, bbox_inches='tight')
```

```python
sns.set_style("ticks")
sns.set_context("talk")

sns.lineplot(x = "coef", y = "dissimilarity", data = df[(df.coef <= 20)], style = "num_iter", hue = "num_iter", palette = 'Paired')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

plt.savefig('./Output/img/dissimilarity-coef_20.png', dpi=1200, bbox_inches='tight')
```
